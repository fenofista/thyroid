# Calculate IOU and DICE score

```python
from ultralytics import YOLO
import numpy as np
import cv2
import glob

# Dice 和 IoU 函數（前面已經寫過）
def dice_score(pred_mask, true_mask):
    intersection = (pred_mask & true_mask).sum()
    return (2. * intersection) / (pred_mask.sum() + true_mask.sum() + 1e-8)

def iou_score(pred_mask, true_mask):
    intersection = (pred_mask & true_mask).sum()
    union = (pred_mask | true_mask).sum()
    return intersection / (union + 1e-8)

# polygon 轉 mask 函數
def polygons_to_mask(polygons, img_shape):
    mask = np.zeros(img_shape, dtype=np.uint8)
    for poly in polygons:
        pts = np.array(poly).reshape((-1, 2))
        pts[:, 0] *= img_shape[1]
        pts[:, 1] *= img_shape[0]
        pts = pts.astype(np.int32)
        cv2.fillPoly(mask, [pts], 1)
    return mask
```

# Train performance

```python
from tqdm import tqdm
model = YOLO("thyroid segmentation/yolo_baseline_aug_space_v2/weights/last.pt")
test_images = glob.glob("../data/YOLO_data/images/train/*.jpg") + glob.glob("../data/YOLO_data/images/train/*.png")   # 改成你的測試集路徑
print(len(test_images))

# ====================================
# 3️⃣ 開始對每張影像做預測 & 比較
# ====================================
dice_scores = []
iou_scores = []
# img_shape = img.shape[:2]  # H, W
min_area_ratio = 0.02  # Example: 1% of the image area

# img_area = img_shape[0] * img_shape[1]

for image_path in tqdm(test_images):
    # 讀取影像
    img = cv2.imread(image_path)
    img_shape = img.shape[:2]  # H, W
    img_area = img_shape[0] * img_shape[1]

    # 預測
    results = model(image_path, imgsz=128, conf = 0.5, verbose = False)
    pred_polygons = []
    for r in results:
        if hasattr(r, 'masks') and r.masks is not None:
            for p in r.masks.xy:
                poly_pts = np.array(p)
                area = cv2.contourArea(poly_pts)
                if area / img_area < min_area_ratio:
                    continue
                # 轉成比例座標（0~1）
                poly = []
                for point in p:
                    x_norm = point[0] / img_shape[1]
                    y_norm = point[1] / img_shape[0]
                    poly.append((x_norm, y_norm))
                pred_polygons.append(poly)

    pred_mask = polygons_to_mask(pred_polygons, img_shape)

    # ====================================
    # 4️⃣ 讀取 Ground truth polygons
    # ====================================
    # 這裡你要根據你的 label.txt 來讀 polygon
    if "jpg" in image_path:
        label_file = image_path.replace("images", "labels").replace(".jpg", ".txt")
    elif "png" in image_path:
        label_file = image_path.replace("images", "labels").replace(".png", ".txt")
    gt_polygons = []
    with open(label_file, 'r') as f:
        for line in f.readlines():
            coords = list(map(float, line.strip().split()[1:]))
            poly = [(coords[i], coords[i+1]) for i in range(0, len(coords), 2)]
            gt_polygons.append(poly)

    gt_mask = polygons_to_mask(gt_polygons, img_shape)

    # ====================================
    # 5️⃣ 計算 Dice 和 IoU
    # ====================================
    dice = dice_score(pred_mask, gt_mask)
    iou = iou_score(pred_mask, gt_mask)

    dice_scores.append(dice)
    iou_scores.append(iou)

    # print(f"{image_path} - Dice: {dice:.4f} - IoU: {iou:.4f}")

# ====================================
# 6️⃣ 輸出整體平均
# ====================================
print(f"Test Set - 平均 Dice: {np.mean(dice_scores):.4f} - 平均 IoU: {np.mean(iou_scores):.4f}")
```
